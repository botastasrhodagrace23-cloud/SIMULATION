```python
# Create an array of 100,000 values from a normal distribution with a mean of 3 and a standard deviation of 1.2
normal_array = np.random.normal(loc=3, scale=1.2, size=100000)
```

```python
import matplotlib.pyplot as plt

# Plot the histogram of these values with a range of -0.6 to +0.6 and a bin width of 0.05
plt.hist(normal_array, bins=np.arange(-0.6, 0.65, 0.05), range=(-0.6, 0.6))
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Normal Distribution")
plt.grid(True)
plt.show()
```

```python
# Create two arrays of 100,000 random integers between 0 and 20
a = np.random.randint(0, 21, 100000)
b = np.random.randint(0, 21, 100000)

# Create array c where c[i] = a[i] * b[i-1]
# The length of c will be 99,999
c = a[1:] * b[:-1]

# Print the length of array c to confirm
print("Length of array c:", len(c))
```

```python
import matplotlib.pyplot as plt

# Create a histogram of array c with a range of -20 to +21
plt.hist(c, bins=np.arange(-20, 22, 1), range=(-20, 21))
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of array c")
plt.grid(True)
plt.show()
```

```python
# Create an array of 100,000 values from a normal distribution with a mean of 0.3 and a standard deviation of 1.2
normal_array = np.random.normal(loc=0.3, scale=1.2, size=100000)

# Print the average and standard deviation of the array
print("Average of the normal array:", np.mean(normal_array))
print("Standard deviation of the normal array:", np.std(normal_array))
```

```python
# Create two arrays of 100,000 random values
random_array_1 = np.random.rand(100000)
random_array_2 = np.random.rand(100000)

# Create a new array by multiplying the two random arrays
multiplied_array = random_array_1 * random_array_2

# Print the average of the multiplied array
print("Average of the multiplied array:", np.mean(multiplied_array))
```

```python
# Create two arrays of 100,000 random values
random_array_1 = np.random.rand(100000)
random_array_2 = np.random.rand(100000)

# Create a new array by summing the two random arrays
summed_array = random_array_1 + random_array_2

# Print the average of the summed array
print("Average of the summed array:", np.mean(summed_array))
```

```python
import numpy as np

# Create an array of 100,000 random values equally distributed between -2 and +3.3
random_array = np.random.uniform(-2, 3.3, 100000)

# Print the lowest and highest values
print("Lowest value:", np.min(random_array))
print("Highest value:", np.max(random_array))
```
